Map = geemap.Map(data_ctrl=False, toolbar_ctrl=False, draw_ctrl=False) digunakan untuk membuat peta interaktif menggunakan pustaka geemap, dengan kontrol data, toolbar, dan kontrol gambar dinonaktifkan.

Kode Map pada baris terakhir digunakan untuk menampilkan peta yang telah dibuat.

```python
Map = geemap.Map(lite_mode=True)
Map
```

Kode Map = geemap.Map(lite_mode=True) digunakan untuk membuat peta interaktif dengan mode ringan (lite mode) menggunakan pustaka geemap. Dalam mode ringan, hanya kontrol zoom in/out yang ditampilkan pada peta.

Kode Map pada baris terakhir digunakan untuk menampilkan peta yang telah dibuat.

```python
Map.save('ipyleaflet.html')
```

Fungsi Map.save('ipyleaflet.html') digunakan untuk menyimpan peta interaktif yang telah dibuat sebagai file HTML dengan nama 'ipyleaflet.html.
Setelah menjalankan kode ini, peta interaktif akan disimpan sebagai file HTML ipyleaflet.html di direktori kerja saat ini. file dapat dibuka di browser web untuk melihat peta.

#2.4 Adding basemaps

2.4.1 Built-in basemaps

```python
import geemap
```

Kode import geemap digunakan untuk mengimpor pustaka geemap dalam Python1geemap adalah paket Python untuk analisis geospasial dan visualisasi interaktif dengan Google Earth Engine (GEE). Dibangun di atas ipyleaflet dan ipywidgets, dan memungkinkan pengguna untuk menganalisis dan memvisualisasikan kumpulan data Earth Engine secara interaktif dalam lingkungan berbasis Jupyter

```python
Map = geemap.Map(basemap='HYBRID')
Map
```

Kode Map = geemap.Map(basemap='HYBRID') digunakan untuk membuat peta interaktif menggunakan pustaka geemap dengan basemap (peta dasar) diatur ke 'HYBRID. Basemap HYBRID biasanya menampilkan peta satelit dengan label jalan dan landmark

```python
Map.add_basemap('OpenTopoMap')
```

Fungsi Map.add_basemap('OpenTopoMap') digunakan untuk menambahkan basemap (peta dasar) OpenTopoMap ke peta interaktif yang telah dibuat menggunakan pustaka geemap. Setelah menjalankan kode ini, basemap OpenTopoMap akan ditambahkan ke peta


```python
for basemap in geemap.basemaps.keys():
 print(basemap)
```

Kode for basemap in geemap.basemaps.keys(): print(basemap) digunakan untuk mencetak semua basemap (peta dasar) yang tersedia dalam pustaka geemap

```python
len(geemap.basemaps)
```

Fungsi len(geemap.basemaps) dalam Python digunakan untuk menghitung jumlah basemap (peta dasar) yang tersedia dalam pustaka geemap.

Terdapat 143 peta dasar yang tersedia dalam pustaka geemap.

2.4.3 WMS tiles

```python
Map = geemap.Map(center=[40, -100], zoom=4)
url = 'https://www.mrlc.gov/geoserver/mrlc_display/NLCD_2019_Land_Cover_L48/wms?'
Map.add_wms_layer(
    url=url,
    layers='NLCD_2019_Land_Cover_L48',
    name='NLCD 2019',
    format='image/png',
    attribution='MRLC',
    transparent=True,
    )

Map
```

Kode Map = geemap.Map(center=[40, -100], url="https://mt1.google.com/vt/lyrs=p&x={x}&y={y}&z={z}", name="Google Terrain", attribution="Google") digunakan untuk membuat peta interaktif dengan pusat di koordinat [40, -100] (lintang 40, bujur -100), dan menggunakan basemap (peta dasar) khusus dari URL yang diberikan.

Kode Map pada baris terakhir digunakan untuk menampilkan peta yang telah dibuat.
